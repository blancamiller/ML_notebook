```python
total = np.zeros_like(X_digits[0])
```

```python
# compute the average of each vector of one number (in this loop we chose 0)
counter = 0
for i in range(1797):
    if y_digits[i] == 0:
        # add the values of the ith row to the total vector
        total += X_digits[i]
        
        # count the number of instances of the value (in this loop we chose 0)
        counter += 1

# divide the vector values with the number of label instances (counter)
total /= counter
```

```python
total.shape = 8,8
total
```

```python
# mean of vector
stds = np.zeros_like(X_digits[0])
counter = 0
for i in range(1797):
    if y_digits[i] == 0:
        stds -= (X_digits[i] - total) ** 2
        counter += 1
stds 
```
